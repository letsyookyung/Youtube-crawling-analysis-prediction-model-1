### 불용어 처리 또는 특수문자제거
- 이 전에 저장해둔 데이터 가져와서, word cloud 형성할 때와 조금 다르게 정리

```python
import konlpy
from konlpy.tag import Okt
import re
```

```python
okt=Okt()
```

- 1차 제거

```python
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)

#분석에 어긋나는 불용어구 제외 (특수문자, 의성어)
han = re.compile(r'[ㄱ-ㅎㅏ-ㅣ!?~,".\n\r#\ufeff\u200d]')

clean_name_lst=[]
for i in range(len(df)):
    a = re.sub(emoji_pattern,'',df['name'].iloc[i])
    
    b = re.sub(han,'',a)
    
    clean_name_lst.append(b)
    
df['clean_name']=clean_name_lst
```

- 2차 제거 
- 워드 클라우드 생성할 때는, 요리사,요리,여러분,이렇게 의미없이 반복되는것을 뺐었는데, 다시 넣어서 원래 정리된 제목으로 정리.
- my_stopwords 랑 다름

```python
my_stopwords = [' \ ','ㄴㄴ','.',')','(','를','을','에','도','ㅋ','SUB','sub','가','과','의','이다','는','★','♥','Eng','ENG','Sub','♡','|',\
                '🔥','~!','/','와','이','로','은','으로','에서','만','개','의','레시피','만레','s','%','↗','[',']','심방','골','주부','이렇게',\
                'ㅏ','으아','악','ㄱ','&','l',':','??','랑','…','!','!!','!!!','!!!!','!!!!!','☆','feat','X','x','TV','다','강슆',\
               '🍜🍜','🎆🎆','🥤🥤🥤','|','🍜','🎂🎂🎂','#','inin','ㅁ','👍👍','CC','🍓🍓🍓','🌲🌲🌲','with','🍓🍓','aHHHH','♥♥♥','@_','@','@@@',\
                'ㅜㅜ','으스스', '스','강쉪','강슆','!^^','!!^^','^^','DongPo','🔥🔥','☕','😍😍😍😍','🍴','😍😍😍','😥😥','◑◑','♥♥♥♥♥♥','!!!!!!','😜😜😜', 'ง', '•̀', '_', '•́ง','👍']
```

```python
clean_word = [] 
for i in df['clean_name']:
    
    review_text = re.sub("[12?~3.,()''->\n]", '', i)
    word_text = okt.morphs(review_text, stem=True)

    if True: word_text = [token for token in word_text if not token in my_stopwords]
        
    clean_word.append(' '.join(word_text))
    
   
df['clean_name'] = clean_word
```

- 명사인 것들만 키워드로 저장

```python
df['clean_name_noun'] = df['clean_name'].apply(lambda x : okt.nouns(x))
```
